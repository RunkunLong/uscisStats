```python
import requests
import matplotlib.pyplot as plt

RECEIPT_NUMBER_BASE = "WAC209009"
RANGE_START = 5000
RANGE_END = 9999
CASE_TYPE = 'I-765'


def checkStatus(ReceiptNum, statusList):
    pload = {"appReceiptNum": ReceiptNum, "initCaseSearch":"CHECK STATUS"}
    r = requests.post('https://egov.uscis.gov/casestatus/mycasestatus.do',data = pload)
    response = r.text.split("\n")
    if CASE_TYPE in response[733]:
        statusList.append((ReceiptNum, response[719].strip()))
        
def savePlot(statusList):
    x, y = zip(*statusList)
    plt.scatter(x, y)
    plt.show()
        
num_digits = len(str(RANGE_END))
statusList = []
for i in range(RANGE_START, RANGE_END):
    ReceiptNum = RECEIPT_NUMBER_BASE + str(i).zfill(num_digits)
    checkStatus(ReceiptNum, statusList)
savePlot(statusList)
```
